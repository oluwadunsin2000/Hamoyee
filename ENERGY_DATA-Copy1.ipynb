```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import statsmodels.api as sn
```

LOADING FILE FROM THIS CSV AND TO VIEW THE SHEETS HEADER

```python
energy_data = pd.read_csv('C:/Users/HP/Downloads/energy_date.csv')
energy_data.head()
```

```python
energy_data.describe()
```

```python
df = energy_data.drop(columns=["date","lights"])
x_multi = df.drop(columns=["Appliances"])
y_multi = df["Appliances"]
```

```python
x_multi_cons = sn.add_constant(x_multi)
x_multi_cons.head()
```

```python
lm_multi = sn.OLS(y_multi, x_multi).fit()
lm_multi.summary()
```

```python
lm3 = LinearRegression()
lm3.fit(x_multi, y_multi)
```

```python
print(lm3.intercept_,lm3.coef_)
```

NOMILIAZING ENERGY DATA

```python
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
features_df = normalised_df.drop(columns=["Appliances"])
Appliances_target = normalised_df["Appliances"]
```

SPLITING DATASET INTO TRAIN AND TEST

```python
x_train, x_test, y_train, y_test = train_test_split(features_df,Appliances_target, test_size=0.3, random_state = 42)
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
predicted_values = linear_model.predict(x_test)
print(linear_model.intercept_,linear_model.coef_)
```

WHAT IS THE MEAN ABSOLUTE ERROR

```python
mae = mean_absolute_error(y_test, predicted_values)
round(mae,2)
```

WHAT IS THE RESIDUAL SUM OF SQUARES

```python
rss = np.sum()
```

WHAT IS THE ROOTS SQUARED ERROR

```python
mse = mean_squared_error(y_test,predicted_values)
round(mse,3)
```

RMSE

```python
rmse = np.sqrt(mse)
round(rmse,3)
```

PENALIZATION METHOD

```python
ridge_reg = Ridge(alpha = 0.4)
ridge_reg.fit(x_train, y_train)
```

```python
predicted_values = ridge_reg.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test,predicted_values))
round(rmse,3)
```

```python
lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(x_train, y_train)
```

```python
predicted_values = lasso_reg.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test,predicted_values))
round(rmse,3)
```
